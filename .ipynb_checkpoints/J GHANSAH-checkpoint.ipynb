```python
# Create a copy of the original DataFrame
new_data = data.copy()

# List of columns you want to check and replace '[x]' with 0
columns_to_check = ['17-Dec', '20-Apr', '20-May', '20-Jun']  # Replace this with your specific columns

# Replace '[x]' with 0 in the specified columns of the new DataFrame
for col in columns_to_check:
    # Ensure the replacement is successful
    new_data[col] = new_data[col].replace('[x]', 0)

# Verify changes by displaying the updated columns
print(new_data[columns_to_check])

#save the new DataFrame to a CSV file, you can use:
new_data.to_csv('updated_data.csv', index=False)
#I replaced the [x] with 0, as it is logical, the dataset has both the [x] and empty datapoints so logically, i replaced the [x] with 0 and below, filled the empty with the mean of the column
```

```python
df=pd.read_csv('updated_data.csv')
```

```python
df.info()
```

```python
# Select all columns starting from the 5th column (index 4)
columns_to_fill = df.columns[4:]

# Fill empty (NaN) values with the mean of each column
for col in columns_to_fill:
    df[col] = df[col].fillna(df[col].mean())

# Save the updated DataFrame as a new DataFrame
new_df = df.copy()

# Display the new DataFrame
print(new_df)
```

```python
#Examine how values change for each region over time.
import seaborn as sns
region_time_data = df.groupby('Region').sum().loc[:, '17-Jan':'24-Aug']
sns.heatmap(region_time_data, cmap="YlGnBu", annot=False)
plt.title('Regional Trends Over Time')
plt.xlabel('Time')
plt.ylabel('Region')
plt.show()
```

```python
high_priority = df[df['Prioritise using SOC 2 as high or low'] == 'high'].loc[:, '17-Jan':'24-Aug'].sum()
low_priority = df[df['Prioritise using SOC 2 as high or low'] == 'low'].loc[:, '17-Jan':'24-Aug'].sum()

print(high_priority.head())
print(low_priority.head())
```

```python
#Explore the differences between "high" and "low" prioritization categories within regions or SOC labels.
# Sum values across columns for high and low prioritization
high_priority = df[df['Prioritise using SOC 2 as high or low'] == 'high'].loc[:, '17-Jan':'24-Aug'].sum()
low_priority = df[df['Prioritise using SOC 2 as high or low'] == 'low'].loc[:, '17-Jan':'24-Aug'].sum()

# Create a stacked area chart
plt.figure(figsize=(20, 12))
plt.stackplot(df.columns[4:], high_priority, low_priority, labels=['High Priority', 'Low Priority'])
plt.legend(loc='upper right')
plt.title('High vs Low Prioritization Over Time')
plt.xlabel('Time')
plt.ylabel('Total Values')
plt.xticks(rotation=45)
plt.show()
```

```python
print(df[df['Prioritise using SOC 2 as high or low'] == 'high'].head())
print(df[df['Prioritise using SOC 2 as high or low'] == 'low'].head())
```

```python
#Compare how SOC 2-digit labels vary across regions.
interaction_data = df.pivot_table(index='Region', columns='SOC 2 digit label', values='Total', aggfunc='sum')
sns.heatmap(interaction_data, cmap="coolwarm", annot=False)
plt.title('Region-SOC Label Interaction')
plt.xlabel('SOC Label')
plt.ylabel('Region')
plt.show()
```

```python
#Analyze time-series changes for individual SOC labels.
selected_label_data = df[df['SOC 2 digit label'] == '11'].loc[:, '17-Jan':'24-Aug']
plt.plot(selected_label_data.columns, selected_label_data.iloc[0])
plt.title('Time Trend for SOC Label 11 (Corporate Managers)')
plt.xlabel('Time')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.show()
```
